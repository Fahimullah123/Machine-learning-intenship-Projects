```python
# Handling missing value or imputing missing value 
from sklearn.impute import SimpleImputer

mean_impute=SimpleImputer(strategy='mean')
df['Life expectancy ']=mean_impute.fit_transform(df[['Life expectancy ']])
df['Adult Mortality']=mean_impute.fit_transform(df[['Adult Mortality']])
median_impute=SimpleImputer(strategy='median')
df['Alcohol']=median_impute.fit_transform(df[['Alcohol']])
df['Polio']=median_impute.fit_transform(df[['Polio']])
median_impute=SimpleImputer(strategy='most_frequent')
df['Total expenditure']=median_impute.fit_transform(df[['Total expenditure']])
df['Hepatitis B']=median_impute.fit_transform(df[['Hepatitis B']])
df['GDP']=median_impute.fit_transform(df[['GDP']])
df['Income composition of resources']=median_impute.fit_transform(df[['Income composition of resources']])
df['Schooling']=median_impute.fit_transform(df[['Schooling']])
df['Population']=median_impute.fit_transform(df[['Population']])
```

```python
df.isnull().sum()
```

```python
# Drop column where is null value
df.dropna(axis=1,inplace=True)
```
